```python
import pandas as pd
import sqlite3
```

```python
# Path to CSV file
csv_file = "D:\\University-Master\\Term-3\\Web Mining\\PRO\\Semantic Search Engine\\Data Cleaning\\output_parsed.csv"

# Read CSV
df = pd.read_csv(csv_file)
```

```python
conn = sqlite3.connect("D:\\University-Master\\Term-3\\Web Mining\\PRO\\Semantic Search Engine\\irandoc.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
SELECT name 
FROM sqlite_master 
WHERE type='table';
""")

tables = cursor.fetchall()

for table in tables:
    print(table[0])
```

```python
cursor.execute("DROP TABLE IF EXISTS documents")
cursor.execute("DROP TABLE IF EXISTS doc")
```

```python
# Save dataframe into SQLite
# df.to_sql("documents", conn, if_exists="replace", index=False)
df.head(100).to_sql("documents", conn, if_exists="replace", index=False)

conn.commit()

print("CSV successfully imported into SQLite.")
```

```python
cursor.execute("PRAGMA table_info(documents);")
columns = cursor.fetchall()

for col in columns:
    print(col)
```

```python
cursor.execute("SELECT * FROM documents LIMIT 3")
rows = cursor.fetchall()

for row in rows:
    print(row)
```
